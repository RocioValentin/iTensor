```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file containing the data
data = pd.read_csv('maxLinkDim_vs_ttotal.csv')

# Plot maxLinkDim vs ttotal
plt.figure(figsize=(8, 6))
plt.plot(data['ttotal'], data['maxLinkDim'], marker='o', linestyle='-', color='b')
plt.xlabel('Total Time (ttotal)')
plt.ylabel('Max Link Dimension')
plt.title('Max Link Dimension vs Total Time')
plt.grid(True)
plt.show()

#int N = 20; //number of sites
# Real tstep = 1; //0.02; //time step (smaller is generally more accurate)
# Real ttotal_max = 10; //1.0; //total time to evolve
# Real cutoff = 1E-8;
```
