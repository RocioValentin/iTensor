```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file containing the data
data = pd.read_csv('maxLinkDim_vs_ttotal.csv')
data10 = pd.read_csv('maxLinkDim_vs_ttotal4.csv')
data15 = pd.read_csv('maxLinkDim_vs_ttotal5.csv')

# Plot maxLinkDim vs ttotal
plt.figure(figsize=(8, 6))
plt.plot(data['ttotal'], data['maxLinkDim'], marker='o', linestyle='-', color='b', label='N = 20')
plt.plot(data10['ttotal'], data10['maxLinkDim'], marker='x', linestyle='--', color='r', label='N = 10')
plt.plot(data15['ttotal'], data15['maxLinkDim'], marker='s', linestyle='-.', color='g', label='N = 15')
plt.xlabel('Total Time (ttotal)')
plt.ylabel('Max Link Dimension')
plt.title('Max Link Dimension vs Total Time')
plt.grid(True)
plt.show()

#int N = 20; 
# Real tstep = 1; 
# Real ttotal_max = 10;
# Real cutoff = 1E-8;
```
